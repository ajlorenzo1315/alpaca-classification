```python
import pandas as pd
import json
import os

# Assuming the CSV has two columns: 'Title' and 'Resultado'

# The 'Title' contains file paths and the 'Resultado' contains the ground truth labels.


# Path to the CSV file

csv_file_path = './0.csv'

# Read the CSV file

df = pd.read_csv(csv_file_path)

# Function to extract the ground truth from the title

def extract_ground_truth(title):

    # The ground truth is the last part of the file path before the extension

    return title.split('/')[-2]
    
def extract_result(result):
    # The ground truth is the last part of the file path before the extension
    return result.split('\n')[0].split(',')[0].split(' ')[-1]

# Apply the function to the 'Title' column to create a new 'ground_truth' column

df['ground_truth'] = df['Titulo'].apply(extract_ground_truth)
df['Resultado_n']= df['Resultado'].apply(extract_result)
# Function to evaluate the classifier

def evaluate_classifier(predictions, ground_truth):

    # Comparing the predicted labels with the ground truth labels

    correct_predictions = (predictions == ground_truth).sum()

    accuracy = correct_predictions / len(ground_truth)

    return accuracy


# Run evaluation
#display(df['Resultado'])
#display(df['Resultado_n'])
#display(df['ground_truth'])
accuracy = evaluate_classifier(df['Resultado_n'][0:3], df['ground_truth'][0:3])
accuracy
```

```python
# Generate the confusion matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(df['ground_truth'], df['Resultado_n'])
# Plotting the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import re
# Path to the CSV file
def data_prepr(result):
     result= result.split(',')[0].replace(" ", "")
     result= result.replace("[", "")
     result= result.replace("]", "")
     result= result.replace("}", "")
     result= result.replace("'", "")
     return result

def find_word_order(text, word_list):
    print(text)
    # Find the index of each word in the text
    indices = {word: text.find(word) for word in word_list}
    # Filter out words that were not found (find returns -1 when it doesn't find the word)
    found_indices = {word: index for word, index in indices.items() if index != -1}
    # Sort the words by their index
    sorted_words = sorted(found_indices, key=found_indices.get)
    return sorted_words


def get_result(result):
    return result.split('result')[-1].split(':')[1].replace("\n", " ")
    
def extract_result(result):
    # The ground truth is the last part of the file path before the extension
    if type(result)==type(''):
        #print(find_word_order(get_result(result),['student','other','staff']))
        #print('text',result)
        result= find_word_order(get_result(result),['staff', 'project', 'other', 'faculty', 'department', 'student', 'course'])
        #print('result',result)
        if len(result)>0:
            return  result[0]
        else:
            return ''
    else:
        return ''
csv_file_path = './resultados_rag1.csv'

# Read the CSV file
df = pd.read_csv(csv_file_path)
df['ground_truth'] = df['Titulo'].apply(extract_ground_truth)

print(list(dict.fromkeys(df['ground_truth'])))

df['Resultado_n']= df['Resultado'].apply(extract_result)
print(df['Resultado'][19])
#print(df['Resultado_n'])
for i,ii in enumerate(df['Resultado_n']):
    print(i,ii)
    
accuracy = evaluate_classifier(df['Resultado_n'], df['ground_truth'])
print(accuracy)
conf_matrix = confusion_matrix(df['ground_truth'], df['Resultado_n'])
# Plotting the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
